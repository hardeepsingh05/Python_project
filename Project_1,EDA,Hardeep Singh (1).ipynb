# 5. Handling of Outliers

```python
e0=data.Engine_size.min()
e100=data.Engine_size.max()
q1=data.Engine_size.quantile(0.25)
q2=data.Engine_size.quantile(0.5)
q3=data.Engine_size.quantile(0.75)
IQR=q3-q1
# IQR - Inter Quartile Range
# q1 - First_quartile
# q2 - Second_quatile
# q3 - Third_quartile
```

```python
lf = q1 - 1.5*IQR
# lc - lower_fence
uf = q3 + 1.5*IQR
# uf - upper_fence
```

```python
lf
```

```python
uf
```

```python
# If lf<e0 then there are NO Outliers on the lower side
# If uf>e100 then there are NO Outliers on the higher side
```

```python
print( "e0 = " , e0 ,", e100 = " , e100 ,", lf = " , lf ,", uf = " , uf)
```

```python
# Let's check the outliers in 'Engine_size' of this Dataset using Boxplot
plt.figure(figsize=(10,5))
sns.boxplot(data=data,x="Engine_size")
plt.show()
```

```python
data.Engine_size.clip(upper=uf)
```

```python
data.Engine_size.clip(upper=uf,inplace=True)
plt.figure(figsize=(10,5))
sns.boxplot(data=data,x="Engine_size")
plt.show()
```

# 6. Handle Missing Values

```python
# Detecting the Missing Values
data.__year_resale_value=pd.to_numeric(data.__year_resale_value,errors='coerce')
data.isnull().sum()
# Dataset.isnull.sum() will give the number of missing values in the dataset
# error='coerce' will replace the non-numeric values with NaN
```

```python
# Missing Values in the Dataset 
# __year_resale_value = 36
# Fuel_efficieny = 3
# all other attributes also have some of the missing values

data.dropna(how='any',inplace=True)
# After this all the missing values are dropped from the dataset
```

```python
data.isnull().sum()
# Here we can check that all the null value counts are zero
```

# 7. Univariate Analysis (Taking only one attribute out of Data)


```python
plt.figure(figsize=(23,10))
sns.countplot(x=data['Manufacturer'],palette='Paired')
plt.show()
# CountPlot will basically display the count of number of the cars made by the Manufacturer
```

```python
# Engine type is measured in Litres(1l to 8l)
plt.figure(figsize=(23,8))
sns.countplot(x=data['Engine_size'],palette='Paired')
plt.show()
```

# 8. Bivariate Analysis (Taking only two attribute out of Data)

```python
# Scatter Plot of 'Engine_size' verse 'Price_of_the_Vehicles'
plt.figure(figsize=(15,8))
sns.scatterplot(data=data,x="Engine_size",y='Price_in_thousands')
plt.show()
```

```python
plt.figure(figsize=(20,8))
sns.barplot(data=data,x='Engine_size',y='Horsepower')
plt.show()
#This barplot is showing 'Horsepower' verse 'Engine_size' of the cars
```

```python
plt.figure(figsize=(25,10))
sns.lineplot(
    data=data, x="Manufacturer", y="Fuel_efficiency", err_style="bars", ci=68
)
plt.show()

# So in this lineplot, we will the 'fuel_efficiency' verse 'Manufacturer' in 'bars' style mode with 'confidence_interval(ci) = 68'
```

```python
sns.displot(data=data, x="Wheelbase", hue="Vehicle_type", multiple="stack")
# Here actually we are trying to make a distplot using Seaborn after giving the 'Vehicle_type' for a bit of colored visualization
```

```python
# In above lineplot we can see that high fuel_efficiency is give by 'Saturn' cars and unfortunately this car company has been discontinued.
# After that 'Chevrolet' cars are performing good and after that 'Hyundai' 
# These bars are telling about the 'fuel_efficiency' range of each car manufacturer.
```

# 9. Multivariate Analysis (Taking multiple attribute out of Data)

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=data,x='Horsepower',y='Fuel_efficiency',palette='bright',hue='Vehicle_type',style='Vehicle_type')
plt.show()
```
